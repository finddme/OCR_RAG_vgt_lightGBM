```python
import ast
import json

# 원본 문자열
original_string = "[{'bbox': [55.0, 506.0, 514.0, 534.0], 'page_number': 1, 'layout': (4, 'List item', (46, 139, 87))}, {'bbox': [70.0, 541.0, 514.0, 567.0], 'page_number': 1, 'layout': (4, 'List item', (46, 139, 87))}, {'bbox': [69.0, 575.0, 513.0, 600.0], 'page_number': 1, 'layout': (4, 'List item', (46, 139, 87))}, {'bbox': [69.0, 607.0, 513.0, 632.0], 'page_number': 1, 'layout': (4, 'List item', (46, 139, 87))}, {'bbox': [255.0, 654.0, 514.0, 710.0], 'page_number': 1, 'layout': (10, 'Text', (0, 0, 255))}, {'bbox': [226.0, 746.0, 380.0, 767.0], 'page_number': 1, 'layout': (7, 'Picture', (255, 0, 255))}, {'bbox': [71.0, 86.0, 521.0, 707.0], 'page_number': 2, 'layout': (9, 'Table', (0, 255, 255))}, {'bbox': [461.0, 61.0, 537.0, 70.0], 'page_number': 3, 'layout': (6, 'Page header', (250, 128, 114))}]"

# 방법 1: ast.literal_eval 사용
try:
    result_ast = ast.literal_eval(original_string)
    print("AST 방법 결과:", result_ast)
except:
    print("AST 방법 실패")

# 방법 2: 문자열 수정 후 json.loads 사용
def fix_json_string(s):
    # 작은따옴표를 큰따옴표로 변경
    s = s.replace("'", '"')
    # 튜플을 리스트로 변경
    s = s.replace("(", "[").replace(")", "]")
    return s

try:
    fixed_string = fix_json_string(original_string)
    result_json = json.loads(fixed_string)
    print("JSON 방법 결과:", result_json)
except json.JSONDecodeError as e:
    print("JSON 파싱 실패:", e)
```
